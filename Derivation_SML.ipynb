### Derivation for SML Equation (Heteroskedastic Dividend Process)


#### First, getting SML equation

Up to now, we have equilibrium price formula we got before, written below:

$$P_i (1+r) = d - \frac{1}{\gamma}\left( b_i \sigma^2_z - \frac{\sigma^2_i}{N} \right) + \mathbb{1}_{i \geq \hat{i}} \frac{\theta}{\gamma} \left(  b_i \sigma^2_z \omega(\lambda) - \frac{\sigma^2_i}{N} \right)\;\;\;\;\;\;\; \text{(1)} $$

$$\text{Where: } \;\;\;\; \omega(\lambda) = \frac{\lambda \gamma - \frac{\sigma_2^z}{N} \sum_{i\geq\bar{i}} b_i}{\sigma^2_z\left( 1 + \sigma^2_z   \sum_{i < \bar{i}} \frac{b_i^2}{\sigma^2_i} \right)}$$

Restating the equilibrium in terms of expected excess returns and relate them to the familiar market $\beta$ from CAPM.

We have now:

$$\tilde{R}_i^e = d + b_i \tilde{z} + \epsilon_i - (1+r)P_i$$

$$\mathbb{E} \left[\tilde{R}_i^e\right] = d - (1+r)P_i$$

$$\tilde{R}_m^e = \sum_{i=1}^N \frac{1}{N} \tilde{R}_i^e =  d + \sum_{i=1}^N \frac{b_i}{N}\tilde{z} + \sum_{i=1}^N \frac{\epsilon_i}{N} -\sum_{i=1}^N(1+r)P_i = d + \tilde{z} + \sum_{i=1}^N \frac{\epsilon_i}{N} - (1+r)P_m$$

$$\mathbb{E} \left[\tilde{R}_m^e\right] = d - (1+r)\sum_{k=1}^N\frac{P_k}{N}$$

Define CAPM beta (stock beta of asset):

$$\beta_i = \frac{cov\left(\tilde{R}_i^e, \tilde{R}_m^e\right)}{var\left(\tilde{R}_m^e \right)} = \frac{b_i \sigma^2_z + \frac{\sigma^2_i}{N}}{\sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2}}$$

Yielding:

$$b_i \sigma^2_z = \beta_i \left( \sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2} \right) - \frac{\sigma^2_i}{N} \;\;\;\;\;\;\; \text{(2)}$$

Substitute formula for $E[R_i^e]$ into price formula $(1)$, we have:

$$\mathbb{E} \left[\tilde{R}_i^e\right]=  \frac{1}{\gamma}\left( b_i \sigma^2_z - \frac{\sigma^2_i}{N} \right) - \mathbb{1}_{i \geq \bar{i}} \frac{\theta}{\gamma} \left(  b_i \sigma^2_z \omega(\lambda) - \frac{\sigma^2_i}{N} \right)$$

Now, substituting $(2)$ into formula above:

$$\mathbb{E} \left[\tilde{R}_i^e\right]=  \frac{1}{\gamma}\beta_i \left( \sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2} \right) - \mathbb{1}_{i \geq \bar{i}} \frac{\theta}{\gamma} \left[  \omega(\lambda) \left[ \beta_i \left( \sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2} \right) - \frac{\sigma^2_i}{N} \right]- \frac{\sigma^2_i}{N} \right]$$

$$\mathbb{E} \left[\tilde{R}_i^e\right]=  \beta_i \frac{\sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2}}{\gamma} \left( 1 - \mathbb{1}_{i \geq \bar{i}} \theta \omega(\lambda) \right) - \theta \frac{\sigma_i^2}{N \gamma} \mathbb{1}_{i \geq \bar{i}} \left( 1 + \omega(\lambda)\right)$$

Or, writing it in piecewise notation, we can remove the indicator function to yield:

$$
 \mathbb{E} \left[\tilde{R}_i^e\right] =
  \begin{cases} 
      \hfill \beta_i \frac{\sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2}}{\gamma}     \hfill & \text{ for $i < \hat{i}$} \\
      \hfill \beta_i \frac{\sigma^2_z + \sum_{k=1}^N \frac{\sigma^2_k}{N^2}}{\gamma} \left( 1 -  \theta \omega(\lambda) \right) - \theta \frac{\sigma_i^2}{N \gamma} \left( 1 + \omega(\lambda)\right) \hfill & \text{for $i \geq \hat{i}$} \\
  \end{cases}
$$

#### Showing that unconstrained portion of SML $i < \hat{i}$ is in fact CAPM: $\beta_i \mathbb{E} \left[\tilde{R}_m^e\right] $

We have, as before:
$$\mathbb{E} \left[\tilde{R}_m^e\right] = d - (1+r) \sum_{k=1}^N \frac{P_k}{N} = \frac{1}{N} \sum_{k=1}^N \left( d - (1+r)P_k \right)$$

Now, subsitution expression for $d - (1+r) P_i$ from price formula $(1)$ when $i < \hat{i}$:

$$\mathbb{E} \left[\tilde{R}_m^e\right] =  \frac{1}{N} \sum_{k=1}^N \frac{1}{\gamma} \left( b_i \sigma^2_z - \frac{\sigma^2_i}{N} \right) = \frac{1}{\gamma} \left( \sigma^2_z \sum_{k=1}^N \frac{b_i}{N} +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right)$$

$$\mathbb{E} \left[\tilde{R}_m^e\right] = \frac{1}{\gamma} \left( \sigma^2_z +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right)$$

Which is the required result.

#### Showing that slope of SML is decreasing with $\lambda$ (aggregate disagreement) after cutoff $i \geq \bar{i}$:

$$Slope = \frac{\partial \mathbb{E} \left[\tilde{R}_i^e\right]}{\partial \beta_i} = \frac{1}{\gamma} \left( \sigma^2_z +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right) (1 - \theta \omega(\lambda))$$
$$ = \text{CAPM Slope} -  \omega(\lambda)\frac{\theta}{\gamma} \left( \sigma^2_z +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right) $$

Now, differentiating w.r.t. $\lambda$:

$$ \frac{\partial^2 \mathbb{E} \left[\tilde{R}_i^e\right]}{\partial \beta_i \partial \lambda}= - \frac{\theta}{\gamma} \left( \sigma^2_z +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right) \frac{\partial \omega(\lambda)}{\partial \lambda} = - \frac{\theta}{\gamma} \left( \sigma^2_z +  \sum_{k=1}^N \frac{\sigma^2_i}{N^2} \right) \frac{\gamma}{\sigma^2_z \left( 1+\sigma^2z \sum_{i<\bar{i}} \frac{b_i^2}{\sigma^2_i} \right)} > 0$$

Since:

- $\alpha > 0$, otherwise no shorting and we dont have constraints (all investors would have same priors/beliefs)
- This implies $\theta > 0$
- $\gamma > 0$, as inverstors are risk averse
- All volatilities/variances are necessarily strictly positive otherwise model is deterministic, which it is not.
